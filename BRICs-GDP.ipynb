```python
from pandas_datareader import wb

matches = wb.search('GDP [current US$]')

dat = wb.download(indicator='NY.GDP.MKTP.CD', country=['BRA', 'CHN', 'IND', 'ZAF', 'RUS'], start=1960, end=2020)
```

```python
pd.set_option('float_format', '{:.0f}'.format) 
```

```python
data = dat.pivot_table(values='NY.GDP.MKTP.CD', index=['year'], columns=['country'])
data
```

```python
data.iplot(kind='line')
```
